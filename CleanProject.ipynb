```python
#Showing all of the data above regarding tenure as a bar chart
#This was done with a lot of assistance from ChatGPT to get it all looking the way I wanted

year = ("2001", "2011", "2021")
tenure = {
    'Own Home': (14916465, 15031914, 15499215),
    'Social Rented': (4157251, 4118461, 4227756),
    'Private Rent/Other': (2586759, 4215669, 5056231),
}
x = np.arange(len(year))
width = 0.3  # Increased width of the bars
multiplier = 0

# Define colors for each category
colors = ['peru', 'goldenrod', 'saddlebrown']

fig, ax = plt.subplots()

for attribute, measurement, color in zip(tenure.keys(), tenure.values(), colors):
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=color)
    # Add value labels for 'Own Home Inc. Loan' within the bar
    if attribute == 'Own Home':
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', rotation=90)
    else:
        # Add value labels for 'Social Rented' and 'Private Rented / Other' above the bars
        ax.bar_label(rects, padding=3, rotation=90, fmt='%d')
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Households')
ax.set_xlabel('Census Year')
ax.set_title('Household Tenure as Found in Censuses from 2001 - 2021')
ax.set_xticks(x + (width * (len(year) - 1) / 2))  # Adjusted position for better alignment
ax.set_xticklabels(year)
ax.set_xlim(-0.5, len(year) + 0.25)
ax.set_ylim(0, 22500000)  # Set y-axis limit to 20,000,000
ax.ticklabel_format(style='plain', axis='y')  # Disable scientific notation for y-axis
ax.legend(loc='upper left', ncol=3)

plt.show()
```
